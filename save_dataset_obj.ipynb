```python
import pandas as pd
import pickle
from trajectory import Dataset
from trajectory import Particle

unpickled_df = pd.read_pickle("objects/SCAN_df.pkl")  
unpickled_df.head(5)
```

```python
dataset = Dataset()

for r in unpickled_df.iterrows():
    name = r[1][0]
    date = r[1][1]
    quantity = r[1][2]
    exposure = r[1][3]
    track_id = r[1][4]
    traj = r[1][5]
    VCL = r[1][6]
    VAP = r[1][7]
    VSL = r[1][8]
    LIN = r[1][9]
    STR = r[1][10]
    WOB = r[1][11]
    BeatCross = r[1][12]
    ALH = r[1][13]
    img = r[1][14]
    agglo_cluster = r[1][15]
    SCAN_cluster = r[1][16]
    SCAN_values = r[1][17]
    
    particle = Particle(name, date, quantity, exposure, track_id, 
                        traj, VCL, VAP, VSL, LIN, STR, WOB, BeatCross, 
                        ALH, img, agglo_cluster, SCAN_cluster, SCAN_values)
    dataset.add_particle(particle)
```

```python
dataset.save()
```
